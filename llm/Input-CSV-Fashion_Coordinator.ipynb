```python
from crewai_tools import WebsiteSearchTool
```

```python
# 🎯 3️⃣ 역할 설정 (Agent 생성)
weather_expert = Agent(
    name="Weather Expert",
    role="날씨 분석가",
    goal="사용자 {request}의 날짜와 위치를 파악하고, 웹 검색을 통해 해당 날짜와 위치에 대한 날씨를 파악한다.",
    backstory="나는 웹 검색을 통해 해당 날짜와 위치에 대한 최신 날씨 정보를 분석하여 사용자가 추위나 더위를 피할 수 있도록 도와준다. 비와 눈 같은 기상 현상에 대비 할 수 있도록 한다.",
    allow_delegation = False,
    verbose = True,
    tool = WebsiteSearchTool()
)

fashion_trend_expert = Agent(
    name="Fashion Trend Expert",
    role="패션 트렌드 분석가",
    goal="웹에서 최신 패션 트렌드를 검색하여 추천 스타일에 반영한다.",
    backstory = "나는 최신 20대 트렌드를 분석하여 트렌드에 맞는 옷을 추천한다. 사용자는 20대로 캐주얼한 스타일을 좋아한다. 캐주얼한 최신 트랜드를 분석한다.",
    allow_delegation = False,
    verbose = True,
    tool = WebsiteSearchTool()
)

stylist = Agent(
    name="Event Stylist",
    role="스타일 리스트",
    goal="{request}에 맞는 옷 스타일을 memory 데이터에 있는 것을 조합하여 추천한다.",
    backstory="나는 날씨 분석가가 제공한 날씨 정보와 패션 트렌드 분석가가 제공한 정보를 바탕으로 날씨, 트렌드, 장소의 분위기를 고려하여 사용자가 적절한 옷차림을 선택할 수 있도록 돕는다. 나는 memory에 있는 사용자의 옷장 데이터를 참고하여 범위 내에서 옷을 추천한다. 옷은 상의, 하의, 겉옷, 신발 4가지로 조합한다. 날씨에 맞춰서 겉옷을 제외할 수 있다.",
    memory = closet_data,
    allow_delegation = False,
    verbose = True
)

# 🎯 4️⃣ 역할에 명령하기 (Task 생성)
task1 = Task(
    description="사용자가 특정 날짜에 방문하는 장소의 날씨를 분석하고 이상 기후나 자연 재해 경보 등이 없는지 확인하기.\n",
    agent=weather_expert,
    expected_output="날씨 분석 결과 (예: 따뜻한 옷, 가벼운 옷), 결과물은 한국어로 생성될 예정입니다."
)

task2 = Task(
    description="웹에서 최신 패션 트렌드를 검색하여 스타일 추천에 반영한다.",
    agent=fashion_trend_expert,
    expected_output="최신 패션 트렌드 요약 및 스타일 추천한다. 결과물은 한국어로 생성될 예정입니다."
)

task3 = Task(
    description="사용자의 방문 장소와 이벤트를 고려하여 적절한 스타일을 추천한다.",
    agent=stylist,
    expected_output="장소 및 행사에 적합한 스타일 제안"
    "마크 다운 형식으로 작성"     
    "결과물은 한국어로 생성될 예정입니다."
)

# 🎯 5️⃣ 팀 결성하기 (Crew 생성)
crew = Crew(
    agents=[weather_expert, fashion_trend_expert, stylist],
    tasks=[task1, task2, task3],
    verbose = 2
)


# 🎯 6️⃣ Crew 실행 (스타일 추천 시작)
result = crew.kickoff(inputs = {"request" : "2025년 2월 12일 서울에 있는 대학교 졸업식에 갈 옷 추천"})
```
